```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.mohfw.gov.in/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
type(soup.findAll('tr'))
rows = soup.findAll('tr')
```

```python
df = pd.DataFrame(columns=['S.No.', 'State/UT', 'ConfirmedIndia', 'ConfirmedForeign', 'Recovered', 'Died'])
i = 0
for row in rows[1:-1]:
    columns = row.find_all('td')
    df_row = [col.text for col in columns]
    print(df_row)
    df.loc[i] = df_row
    i = i+1
```

```python
oldDF = pd.read_json('./Data.json', orient='split')
oldDF
```

```python
mergedStuff = pd.merge(df, oldDF, on=['State/UT'], how='outer')
mergedStuff
```

```python
def isSame(row):
#     print(str(row['ConfirmedIndia_x']) == str(row['ConfirmedIndia_y']))
#     print(str(row['ConfirmedForeign_x']) == str(row['ConfirmedForeign_y']))
#     print(str(row['Recovered_x']) == str(row['Recovered_y']))
#     print(str(row['Died_x']) == str(row['Died_y']))
    result = float(row['ConfirmedIndia_x']) == row['ConfirmedIndia_y'] and float(row['ConfirmedForeign_x']) == row['ConfirmedForeign_y'] and float(row['Recovered_x']) == row['Recovered_y'] and float(row['Died_x']) == row['Died_y']
#     print(result)
    return result
indexes = []
for index, row in mergedStuff.iterrows():
    if isSame(row):
        indexes.append(index)

mergedStuff = mergedStuff.drop(indexes)
print(indexes)
```

```python
if not mergedStuff.empty:
    heading = ["State", "CI(now)", "CI(earlier)", "CF(now)", "CF(earlier)", "Rec(now)", "Rec(earlier)", "Deaths(now)", "Deaths(earlier)"]
    stats = []
    for index, row in mergedStuff.iterrows():
        stats.append([row["State/UT"], row["ConfirmedIndia_x"], row["ConfirmedIndia_y"], row["ConfirmedForeign_x"], row["ConfirmedForeign_y"], row["Recovered_x"], row["Recovered_y"], row["Died_x"], row["Died_y"]])
    heading
    stats
    # stats = heading + stats
    text = tabulate(stats, headers=heading, tablefmt="orgtbl")
else:
    print("empty")
```
